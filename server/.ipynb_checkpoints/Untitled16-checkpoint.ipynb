```python
#1
doc=nlp('I (go) to school all the afternoon')
def getsentencetense(doc,tense_uses):
    material_patterns = [nlp(text) for text in tense_uses]
    matcher = PhraseMatcher(nlp.vocab)
    matcher.add('tense', None, *material_patterns)
    matches = matcher(doc)
    return len(matches)!=0

# getsentencetense(doc,past_continous_uses)
```

```python
#2
def getpronouns(text):
    result = re.sub(r'\'m', r' am', text)    
    result = re.sub(r'\'s', r' is', result)  
    result = re.sub(r'\'re', r' are', result) 
    result = re.sub(r'n\'t', r' not', result)    
    return(result)
```

```python
#3
def replacenounwithpronoun(text):
    verbs = re.findall(r"\((\w+)\)", text)
    text = re.sub(r'\(|\)', r'', text)
    doc = nlp('{}'.format(text))
    
    for [noun,pronoun] in doc._.coref_clusters:
        doc = re.sub(r'{}'.format(noun), r'{}'.format(pronoun), str(doc))
    
    for verb in verbs:
        doc=re.sub(r'{}'.format(verb), r'({})'.format(verb), str(doc))      
    return doc

# replacenounwithpronoun('sara has a dog and she love him')
```

```python
#4
def getsubjectandverb(text):
    doc = nlp1('{}'.format(text))
    sentences = []
    sentences = re.split(r', | and | or | but | so | nor ', str(doc))
    dic={}
    pronoun=0
    for sent in sentences:
        verb = re.search(r"\((\w+)\)", sent)
        sent = re.sub(r'\(|\)', r'', sent)
        sent=nlp1('{}'.format(sent))
        displacy.render(sent, style="dep")
        if verb !=None:
            verb = verb.group(1)
            for word in sent:
                if word.pos_=='PRON' and word.head.pos_ == 'VERB'and word.head.text == verb or re.search(r'{}'.format(word.text),str(pronouns))!=None and word.pos_=='PRON' and word.head.pos_ == 'NOUN'and word.head.dep_=='nsubj':
                    pronoun=word 
                    break
            dic[pronoun]=verb
    return dic

getsubjectandverb('she (love) traveling and She (go) abroad almost every summer')
```

```python
#5
def prepare(doc):
    tools={}
    doc=getpronouns(str(doc))
    doc=replacenounwithpronoun(str(doc))
    tools=getsubjectandverb(str(doc))
    dic=dict(tools)
    return dic
```

```python
def verb_ing(token):
    special_verbs=nlp1('age agree disagree dye free knee refere singe tiptoe')
    # if it is a uncontinuous verbs
    if re.search(r' {} '.format(token),str(non_continuous_verb))!=None: return token
    #special case
    if re.search(r' {} '.format(token),str(special_verbs))!=None: return token+"ing"
    #lie-->lying
    if token[-2:]=='ie': return token[:-2] +'ying'
    # live --> living
    if token[-1]=='e': return token[:-1]+"ing"
    #fix --> fixing, enjoy --> enjoying, snow --> snowing
    if token[-1] in ['w','x','y'] and token[-2] in vowel: return token + "ing"
    #stop-->stopping
    if token[-2] in ["u",'o','i','a']: return token +token[-1]+"ing"
    return token +'ing'

verb_ing('lie')
```

```python
def verb_s(token):
    if token[-1]=='y' and re.search(token[-2],str(vowel)) != None: return token+"s"
    #cry --> cries
    if token[-1]=='y':return token[:-1]+"ies" 
    #watch --> watches, guess --> guesses, mix --> mixes
    if token[-2:]in ['ss','ch','zz','sh'] or token[-1] in['x','o'] or token[-3:]=="tch" : return token+"es"
    if token.lower()=='have': return "has"
    if token.lower()=='be':  return "is"
    return token+"s"
verb_s('play')
```

```python
def verb_ed(token): 
    #play-->played , box-->boxed ,row-->rowed
    if token[-1] in ['w','x','y']  and re.search(token[-2],str(vowel)) != None: return token +"ed"
    #study-->studied
    if token[-1]=='y' and re.search(token[-2],str(consonant)) != None: return token[:-1]+"ied"
    #arrive-->arrived
    if token[-1]=='e': return token +"d"
    #stop-->stopped
    if re.search(token[-1],str(consonant))!= None and re.search(token[-2],str(vowel)) != None: return token + token[-1]+ed
    #have,has-->had
    if token.lower()=='have' or token.lower()=='has': return "had" 
    return token +"ed"
verb_ed('arrive')
```

```python
def present_simple(doc):
    dic=prepare(doc)
    for i in dic:
        verb=dic[i]
        if re.search(r'{}'.format(i),'he she it He She It') != None:
            doc = re.sub(r'{}'.format(verb), r'{}'.format(verb_s(verb)), str(doc))      
        elif re.search(r'{}'.format(i),'I') != None and verb=='be':
            dic[i]='am'
        elif verb=='be':
            dic[i]='are'
        doc = re.sub(r'\(|\)',r'',str(doc))
    return doc
present_simple('she (play) games, she is good')
```

```python
# elif token.lower()=='be':
#         token='was'
#         return token

def past_simple(doc):
    dic=prepare(doc)
    for i in dic:
        verb=dic[i] 
        if re.search(r"{}".format(verb), str(unregular_verbs)) != None:
            verb = re.search(r"{}".format(verb), str(unregular_verbs))
            verb=verb.group(0)
            for k,v in dict_v2.items():
                if k==verb:
                    dic[i]=v
                    doc = re.sub(r'{}'.format(verb), r'{}'.format(dic[i]), str(doc)) 
                    doc = re.sub(r'\(|\)',r'',str(doc))
                else:
                    pass
        else:
            dic[i]=verb_ed(verb)
            doc = re.sub(r'{}'.format(verb), r'{}'.format(dic[i]), str(doc))
            doc = re.sub(r'\(|\)',r'',str(doc))
    return doc

past_simple('adam (win) another prize in 1999 and he is very proud ')
```

```python
def present_continuous(doc):
    dic=prepare(doc)
    for i in dic:
        verb=dic[i]
        if re.search(r'{}'.format(i),'he she it He She ')!=None:
            doc = re.sub(r'{}'.format(verb), r'is {}'.format(verb_ing(dic[i])), str(doc))  
            doc = re.sub(r'\(|\)',r'',str(doc))

        elif re.search(r'{}'.format(i),' they we you They We You ')!=None:
            doc = re.sub(r'{}'.format(verb), r'are {}'.format(verb_ing(dic[i])), str(doc)) 
            doc = re.sub(r'\(|\)',r'',str(doc))
                
        else:
            doc = re.sub(r'{}'.format(verb), r'am {}'.format(verb_ing(dic[i])), str(doc))  
            doc = re.sub(r'\(|\)',r'',str(doc))
     
    return doc
present_continuous('I (drink) my tea now')
```

```python
def past_continuous(doc):
    dic=prepare(doc)
    for i in dic:
        verb=dic[i]
        if re.search(r'{}'.format(i),' he she it I He She It ')!=None:
            doc = re.sub(r'{}'.format(verb), r'was {}'.format(verb_ing(dic[i])), str(doc))  
            doc = re.sub(r'\(|\)',r'',str(doc))

        elif re.search(r'{}'.format(i),'they we you They We You ')!=None:
            doc = re.sub(r'{}'.format(verb), r'were {}'.format(verb_ing(dic[i])), str(doc)) 
            doc = re.sub(r'\(|\)',r'',str(doc))

    return doc

past_continuous('sara (write) another book during the night and she fall a sleep')
```

```python
def present_perfect(doc):
    dic=prepare(doc)
    for i in dic:
        verb=dic[i]
        if re.search(r"{}".format(verb), str(unregular_verbs)) != None:
            verb = re.search(r"{}".format(verb), str(unregular_verbs))
            verb=verb.group(0)
            for k,v in dict_v2.items():
                if k==verb:
                    v2=v
                    for k_1,v_1 in dict_v3.items():
                        if k_1==v2:
                            dic[i]=v_1
                            if re.search(r'{}'.format(i),' he she it I He She It ')!=None:
                                doc = re.sub(r'{}'.format(verb), r' has {}'.format(dic[i]), str(doc)) 
                                doc = re.sub(r'\(|\)',r'',str(doc))
                            else:
                                doc = re.sub(r'{}'.format(verb), r' have {}'.format(dic[i]), str(doc)) 
                                doc = re.sub(r'\(|\)',r'',str(doc))
                        else:
                            pass
        else:
            dic[i]=verb_ed(verb)
            if re.search(r'{}'.format(i),' he she it I He She It ')!=None:
                doc = re.sub(r'{}'.format(verb), r'has {}'.format(dic[i]), str(doc))
                doc = re.sub(r'\(|\)',r'',str(doc))
            else:
                doc = re.sub(r'{}'.format(verb), r'have {}'.format(dic[i]), str(doc))
                doc = re.sub(r'\(|\)',r'',str(doc))
  
    return doc

present_perfect('She (go) abroad just')
```

```python
def past_perfect(doc):
    dic=prepare(doc)
    for i in dic:
        verb=dic[i]
        if re.search(r"{}".format(verb), str(unregular_verbs)) != None:
            verb = re.search(r"{}".format(verb), str(unregular_verbs))
            verb=verb.group(0)
            for k,v in dict_v2.items():
                if k==verb:
                    v2=v
                    for k_1,v_1 in dict_v3.items():
                        if k_1==v2:
                            dic[i]=v_1
                            doc = re.sub(r'{}'.format(verb), r'had {}'.format(dic[i]), str(doc)) 
                            doc = re.sub(r'\(|\)',r'',str(doc))
                        else:
                            pass
        else:
            dic[i]=verb_ed(verb)
            doc = re.sub(r'{}'.format(verb), r'had {}'.format(dic[i]), str(doc))
            doc = re.sub(r'\(|\)',r'',str(doc))
    
    return doc

past_perfect('I (finish) ')
```

```python
def future(doc):
    dic=prepare(doc)
    for i in dic:
        verb=dic[i]
        doc = re.sub(r'{}'.format(verb), r'will {}'.format(verb), str(doc))      
        doc = re.sub(r'\(|\)',r'',str(doc))     
    return doc

future('i (sing) another new song soon')
```

```python
def continuous_future(doc):
    dic=prepare(doc)
    for i in dic:
        verb=dic[i]
        verb=verb
        doc = re.sub(r'{}'.format(verb), r'will be {}'.format(verb_ing(dic[i])), str(doc))  
        doc = re.sub(r'\(|\)',r'',str(doc))       
    return doc

continuous_future('i (sing) another new song soon')
```

```python
def perfect_future(doc):
    dic=prepare(doc)
    for i in dic:
        verb=dic[i]
        if re.search(r"{}".format(verb), str(unregular_verbs)) != None:
            verb = re.search(r"{}".format(verb), str(unregular_verbs))
            verb=verb.group(0)
            for k,v in dict_v2.items():
                if k==verb:
                    v2=v
                    for k_1,v_1 in dict_v3.items():
                        if k_1==v2:
                            dic[i]=v_1
                            doc = re.sub(r'{}'.format(verb), r'will have {}'.format(dic[i]), str(doc)) 
                            doc = re.sub(r'\(|\)',r'',str(doc))
                        else:
                            pass
        else:
            dic[i]=verb_ed(verb)
            doc = re.sub(r'{}'.format(verb), r'will have {}'.format(dic[i]), str(doc))
            doc = re.sub(r'\(|\)',r'',str(doc))
    return doc

perfect_future('i (write) another new song soon')
```

```python
def Present_perfect_continuous(doc):
    dic=prepare(doc)
    for i in dic:
        verb=dic[i]
        if re.search(r'{}'.format(i),' he she it I He She It ')!=None:
            doc = re.sub(r'{}'.format(verb), r'has been {}'.format(verb_ing(dic[i])), str(doc)) 
            doc = re.sub(r'\(|\)',r'',str(doc))
        else:
            doc = re.sub(r'{}'.format(verb), r'have been {}'.format(verb_ing(dic[i])), str(doc)) 
            doc = re.sub(r'\(|\)',r'',str(doc))

    return doc

Present_perfect_continuous('I (work) for this company for more than thirty year')
```

```python
def Past_perfect_continuous(doc):
    dic=prepare(doc)
    for i in dic:
        doc = re.sub(r'{}'.format(dic[i]), r'had been {}'.format(verb_ing(dic[i])), str(doc)) 
        doc = re.sub(r'\(|\)',r'',str(doc))
       
    return doc

Past_perfect_continuous('I (sing) another new song soon')
```

```python
tenses = [
 {
    "uses": ["in the present","every","sometimes","always","often","usually","seldom","never","normally","traditionally","now"],
    "fun": present_simple
  },
  {
    "uses": ["in the past", "yesterday", "ago", "the other day"],
    "fun": past_simple
  },
  {
    "uses": ["soon", "in the future", "in one year", "in a year", "in year"],
    "fun": future
  },
   {
    "uses": ["at the moment","look!","listen!"," next week "," next month "," next year ","next day ","permanently","regularly"],
    "fun": present_continuous
  },
 {
    "uses": ["by this time tomorrow","by this time next week","by this time next year","by this time year",
      "at this time tomorrow","at this time next month","by 3 o’clock tomorrow",
      "by this time next summer","by this time next autumn","by this time next winter","by this time next spring"],
    "fun": continuous_future
  },
 {
    "uses": [
      "by the end of this year","by the end of tomorrow","by the end of this term","by the end of this week","by the end of this month",
      "in another years ","in another months","in another days","in another weeks"
    ],
    "fun": perfect_future
  },
{
    "uses": ["as long as","during","all the night","all the day","at night","all the afternoon","all the morning"],
    "fun": past_continuous
  },
   {
    "uses": [
      "just","yet","never","ever","already","so far","up to now","recently","today",
      "this week","this month","this year","this day","lately","once","twice","many times","several times"],
    "fun": present_perfect
  },
  {
    "uses": [
      "already","by the time","as soon as",'by the end of the year'],
    "fun": past_perfect
  },
  {
    "uses": ["all day", "all night", "all week", "still"],
    "fun": Past_perfect_continuous
  },
  {
    "uses": [' for ',' since '],
    "fun": Present_perfect_continuous
  }
    
];


special_words=nlp1('while when before after')
```

```python
def special_cases(doc):
    sents=[]
    
    u=re.search(r'util',str(doc))  
    if u!=None:
        sents=re.split(r'(util)',str(doc))
        doc=nlp('{} util {}'.format(past_perfect(nlp(sents[0])),past_simple(nlp(sents[2]))))
        return doc

    a=re.search(r'after',str(doc))
    if a!=None:
        sents=re.split(r'(after)',str(doc)) 
        doc=nlp('{} after {}'.format(past_simple(nlp(sents[0])),past_perfect(nlp(sents[2]))))
        return doc
    
    b=re.search(r'before',str(doc))
    if b!=None:
        sents=re.split(r'(before)',str(doc))
        doc=nlp('{} before {}'.format(future(nlp(sents[0])),present_simple(nlp(sents[2]))))
        return doc

    when=re.search(r'when',str(doc))
    if when!=None:
        sents=re.split(r'(when)',str(doc)) 
        doc=nlp('{} when {}'.format(Past_perfect_continuous(nlp(sents[0])),past_continuous(nlp(sents[2]))))
        return doc

    wh=re.search(r'while',str(doc))
    if wh!=None:
        sents=re.split(r'(while)',str(doc))  
        doc=nlp('{} while {}'.format(past_continuous(nlp(sents[0])),past_continuous(nlp(sents[2]))))
        return doc
    
    be=re.search(r'because',str(doc))
    if be!=None:
        sents=re.split(r'(because)',str(doc))  
        doc=nlp('{} because {}'.format(past_simple(nlp(sents[0])),past_continuous(nlp(sents[2]))))
        return doc

special_cases('I (go) to school when I (finish)')
```

```python
def GarmmerSolver(text):
    doc = nlp(u'{}'.format(text)) 
    if re.search(r' before|after|util|when|while|because',str(doc))!=None:   return special_cases(doc)   
    for ent in doc.ents:
        if ent.label_=='NUM' and re.search('in {} year'.format(ent.text),str(doc))!= None:return(future(doc))            
    for ent in doc.ents:
        if ent.label_=='DATE' and re.search('last {}'.format(ent.text),str(doc))!= None:return(past_simple(doc))       
    if re.search('in \d{4}',str(doc))!= None:return(past_simple(doc)) 
    if re.search('by \d{4}',str(doc))!= None:return(past_perfect(doc))    
    if re.search('at \d{1} pm|am',str(doc))!= None: return(past_continuous(doc))
    
    for t in tenses:
        if(getsentencetense(doc,t['uses'])):
            print(t['fun'])
            return t['fun'](doc)
    return(present_simple(doc))
```
